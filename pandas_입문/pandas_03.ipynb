## 4개의 그래프 만들기
- 1. 전체 그래프가 위치할 기본 틀을 만듭니다.
- 2. 그래프를 그려 넣을 그래프 격자를 만듭니다.
- 3. 그런 다음 격자에 그래프를 하나씩 추가합니다. 격자에 그래프가 추가되는 순서는 왼쪽에서 오른쪽 방향
- 4. 만약 격자의 첫번째 행이 꽉차면 두번째 행에 그래프를 그려 넣습니다.

```python
# data 값 저장
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']
```

```python
# 그래프 격자가 위치할 기본 틀을 만듬
fig = plt.figure()
```

```python
# add_subplot 메서드로 그래프격자를 그림
# (기본틀의 행크기, 열크기, 위치)
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

```python
# plot 메서드에 데이터를 전달
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

fig
```

# matplotlib 라이브러리 자유자재로 사용하기

```python
# matplotlib 라이브러리 자유자재로 사용하기
tips = sns.load_dataset("tips")
print(tips)
print(type(tips))
```

```python
# 변수 하나만 사용해서 만든 그래프 : 일변량 그래프
# hist 메서드에 total_bill 열을 전달
fig = plt.figure() # 기본틀 마련
axes1 = fig.add_subplot(1,1,1)
axes1.hist(tips['total_bill']) #bin=10) # bins 인자값은 x 축의 간격을 조정
axes1.set_title('histogram')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

## 산점도 그래프 그리기

```python
# 산점도 그래프 그리기
# 변수 2개를 사용하므로 '이변량 그래프'라고 부름
scatter_plot = plt.figure()
axes2 = scatter_plot.add_subplot(1,1,1)
axes2.scatter(tips['total_bill'],tips['tip'])

axes2.set_title('Bill vs Tip')
axes2.set_xlabel('Total Bill')
axes2.set_ylabel('Tip')
```

## 박스 그래프
- 이산형 변수와 연속형 변수를 함께 사용하는 그래프
    - 이산형 변수란 Male, Female과 같이 명확하게 구분되는 값을 의미
    - 연속형 변수란 Tip과 같이 명확하게 셀수 없는 범위의 값

```python
box_plot = plt.figure()
axes3 = box_plot.add_subplot(1,1,1)
axes3.boxplot([ tips[tips['sex']=='Male']['tip'],
                tips[tips['sex']=='Female']['tip']],
                labels=['Male','Female'])
axes2.set_title('Boxplot of Tips by Sex')
axes2.set_xlabel('Sex')
axes2.set_ylabel('Tip')
```

## 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프

```python
# 각 문자열을 정수로 치환하는 함수 생성
def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
```

```python
# 함수가 반환한 값을 데이터 프레임 sex_color 열을 새로 생성하여 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes2 = scatter_plot.add_subplot(1,1,1)
axes2.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10,
    c = tips['sex_color'],
    alpha = 0.5
    )

axes2.set_title('Bill vs Tip')
axes2.set_xlabel('Total Bill')
axes2.set_ylabel('Tip')
```

# seaborn 라이브러리 사용하기

```python
tips = sns.load_dataset("tips")
```

```python
# seaborn 라이브러리로 히스토그램을 그리려면 subplot, displot 메서드를 사용한다.
# 1. subplot로 기본 틀을 만들고 displot에 데이터를 전달한다.
ax = plt.subplot()
ax = sns.displot(tips['total_bill'])
```

```python
ax = plt.subplot()
# kind가 hist인 경우 히스토그램 출력
ax = sns.displot(tips['total_bill'], kind="hist")
```

```python
ax = plt.subplot()
# kind가 kde인 경우 밀집도 그래프 출력
ax = sns.displot(tips['total_bill'],kind='kde')
```
