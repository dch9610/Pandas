```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 데이터 가져오기
tips = sns.load_dataset('tips')
tips
```

## 데이터프레임과 시리즈로 그래프그리기

```python
# 히스토그램 그리기 hist()
ax = plt.subplot()
ax = tips['total_bill'].plot.hist()
```

```python
# 히스토그램 투명도 조절 (alpha = 투명도, bins = 데이터 등분수, ax = 나타낼 데이터)
fig, ax = plt.subplots()
ax = tips[['total_bill','tip']].plot.hist(alpha=0.5,bins= 20, ax=ax)
```

```python
# 밀집도 그래프 (kde)
kde = plt.subplot()
kde = tips['tip'].plot.kde()
```

```python
# 산점도 그래프 (scatter)
fig, sca = plt.subplots()
sca = tips.plot.scatter(x='total_bill', y = 'tip', ax=sca)
```

```python
# 육각 그래프 (hexbin)
# 크기 변경 (,gridsize)
fig, hex = plt.subplots()
hex = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=hex)
```

```python
# 박스 그래프 (box)
fig, box = plt.subplots()
box = tips.plot.box(ax=box)
```

### seaborn 라이브러리로 그래프 스타일 설정

```python
# 그래프에 스타일 적용하기
# 배경에 가로줄 생기게 하기
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
# for문 이용하여 모든 스타일 적용해보기
fig = plt.figure()
seaborn_styles = ['darkgrid', 'whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2,3,plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data = tips, ax=ax )
        violin.set_title(style)
fig.tight_layout()
```
